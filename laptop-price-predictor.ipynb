```python
def fetch_processor(text):
    if text =='Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'other intel Processor'

        if text.split()[0] == 'Apple':
            return 'Apple Processor'
        else:
            return 'AMD processor'
        
```

```python
df['Cpu brand'] = df['Cpu Name'].apply(fetch_processor)
```

```python
df['SSD'] = 0
df['HDD'] = 0
df['NVIDIA GeForce MX350 2GB GDDR5 VRAM'] = 0

# Split the memory column and update SSD and HDD columns accordingly
for index, row in df.iterrows():
    Memory = row['Memory']
    if 'SSD' in Memory:
        df.at[index, 'SSD'] = Memory.split('SSD')[0].strip()
    elif 'HDD' in Memory:
        df.at[index, 'HDD'] = Memory.split('HDD')[0].strip()
    elif 'NVIDIA GeForce MX350 2GB GDDR5 VRAM' in Memory:
        df.at[index, 'NVIDIA GeForce MX350 2GB GDDR5 VRAM'] = Memory.split('NVIDIA GeForce MX350 2GB GDDR5 VRAM')[0].strip()

# Remove non-numeric characters from SSD and HDD columns
df['SSD'] = df['SSD'].str.replace('\D', '', regex=True)
df['HDD'] = df['HDD'].str.replace('\D', '', regex=True)
df['NVIDIA GeForce MX350 2GB GDDR5 VRAM'] = df['NVIDIA GeForce MX350 2GB GDDR5 VRAM'].str.replace('\D', '', regex=True)

# Convert the SSD and HDD columns to numeric data type
df['SSD'] = pd.to_numeric(df['SSD'], errors='coerce')
df['HDD'] = pd.to_numeric(df['HDD'], errors='coerce')
df['NVIDIA GeForce MX350 2GB GDDR5 VRAM'] = pd.to_numeric(df['NVIDIA GeForce MX350 2GB GDDR5 VRAM'], errors='coerce')

# Fill NaN values with 0
df['SSD'].fillna(0, inplace=True)
df['HDD'].fillna(0, inplace=True)
df['NVIDIA GeForce MX350 2GB GDDR5 VRAM'].fillna(0, inplace=True)
```
